```python
from scripts.utils.simple_qa import load_simple_qa 

DEST = '../../data/preprocessed_question_relation_test.tsv'

df, = load_simple_qa(test=True)
df[:5]
```

```python
from scripts.utils.link_subject_name import add_subject_name

add_subject_name(df, cursor, print_=True)
```

```python
from tqdm import tqdm_notebook
import random
from scripts.utils.link_subject_name import tokenize

preprocessed = [] 
for index, row in tqdm_notebook(df.iterrows(), total=df.shape[0]):
    if isinstance(row['subject_name'], str):
        ret =  ''
        for i, token in enumerate(tokenize(row['question'])):
            if i == row['subject_name_start_index']:
                ret += '<e>'
            elif i > row['subject_name_start_index'] and i < row['subject_name_stop_index']:
                continue
            else:
                ret += token.lower()
            ret += ' '

        preprocessed.append({
            'question': ret.strip(),
            'relation': row['relation'],
        })

random.sample(preprocessed, 50)
```

```python
import pandas as pd
df = pd.DataFrame(preprocessed)
df[:5]
```

```python
import csv 

print('Saved in:', DEST)
df.to_csv(DEST, sep='\t', quoting=csv.QUOTE_NONE, index=False)
```
